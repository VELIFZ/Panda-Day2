### Homework Excersise #1 - Find the Mean for the Red Sox <br>
<p>Get total number of home runs and rbi's</p>

```python
bs17 = pd.read_csv('../../Files/redsox_2017_hitting.txt') #  by default sep = ","
bs18 = pd.read_csv('../../Files/redsox_2018_hitting.txt')
# step 1: Add a new column with the key 'Team' and all column values should be 'BOS'
bs17['Team'] = 'BOS'
bs18['Team'] = 'BOS'
# step 2: Group by the 'Team' column and get total home runs and rbi's - sum() aggregate function
bs17 = bs17.groupby('Team').sum()[['HR', 'RBI']]
bs18 = bs18.groupby('Team').sum()[['HR', 'RBI']]
# Produce data for both 2017 and 2018
print(bs17)
print(f'\n{bs18}')
```
